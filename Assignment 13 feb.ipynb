```python
Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions
Ans:When creating a custom exception, we need to inherit from the Exception class because it provides the basic functionality required by all exceptions
, such as message passing and stack trace tracking. Inheriting from Exception ensures that the custom exception behaves like a standard exception and 
can be caught by a try-catch block, logged, or handled in a consistent way with other exceptions in the codebase. It also allows the custom exception to
leverage any additional features or functionality provided by the Exception class or any of its subclasses.
```

```python
 # Write a python program to print Python Exception Hierarchy.
import logging
logging.basicConfig(filename= "test.log",level = logging.INFO)
def print_exception_hierarchy(exceptions, level=0):
    for exception in exceptions:
        print('\t' * level, exception.__name__)
        logging.info(print_exception_hierarchy(exception.__subclasses__(), level+1))

print_exception_hierarchy([Exception])
```

```python
# What errors are defined in the ArithmeticError class? Explain any two with an example
# The ArithmeticError class is a built-in Python exception class that serves as the base class for several other built-in 
# exception classes that represent specific types of arithmetic errors. Some of the exceptions that are defined in the ArithmeticError
# class are:

#     FloatingPointError: This exception is raised when a floating-point calculation results in an error.
#     For example, when dividing by zero in a floating-point operation.

#     OverflowError: This exception is raised when a calculation exceeds the maximum representable value of a numeric data type.

#     ZeroDivisionError: This exception is raised when an attempt is made to divide a number by zero.

#     UnderflowError: This exception is raised when a calculation results in a value that is too small to be represented by a numeric data type.

#     ArithmeticError: This is the base class for all arithmetic exceptions.

# All of these exceptions are related to arithmetic operations and represent different types of errors that can occur during such operations.
import logging
logging.basicConfig(filename= "test.log",level = logging.ERROR)
try:
    result=5/0
except ZeroDivisionError as e:
    logging.error("Zerodivision Error Occor")

# FloatingPointError
import math
try:
    x = math.sqrt(-1)
except FloatingPointError as e:
    logging.error("Caught FloatingPointError:")
```

```python
# Why LookupError class is used? Explain with an example KeyError and IndexError
# Ans :The LookupError class is a built-in Python exception class that serves as the base class for several other
# built-in exception classes that represent errors related to key or index lookups in data structures.
import logging
logging.basicConfig(filename= "test.log",level = logging.ERROR)
# example:Keyerror
try:
    my_dict = {"apple": 1, "banana": 2, "orange": 3}
    my_dict["grape"]
except KeyError as e:
    logging.error("KeyError{}".format(e))
# IndexError
try:
    my_list = [1, 2, 3]
    my_list[3]
except IndexError as e:
    logging.error("Indexerror{}".format(e))
```

```python
# Explain ImportError. What is ModuleNotFoundError?
# ImportError is a built-in Python exception class that is raised when there is an error while importing a module. 
# This can happen when a module does not exist, cannot be found in the search path, or when there is a syntax error in the module being imported.
# ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is raised when a module could not 
# be found in the search path. In other words, when the Python interpreter is unable to locate the module that is being imported.
import logging
logging.basicConfig(filename= "test.log",level = logging.ERROR)
# This will raise a ModuleNotFoundError
try:
    import my_module
except ModuleNotFoundError as e:
    logging.error("Caught ModuleNotFoundError:{}" .format(e))

# This will raise an ImportError
try:
    import syntax_error_module
except ImportError as e:
    logging.error("Caught ImportError:{}" .format(e))
```
