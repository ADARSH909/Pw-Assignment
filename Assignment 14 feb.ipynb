```python
what is multithreading in python? why is it used? Name the module used to handle threads in python
Ans:
Multithreading in Python is a way of achieving parallelism and concurrency by having multiple threads of execution within a single process. 
Each thread operates independently, and multiple threads can run concurrently on a multi-core CPU, allowing for more efficient use of system
resources.

Multithreading is used in Python when we need to perform multiple tasks simultaneously or when we want to take advantage of multiple cores
in a CPU. It is commonly used in applications that require concurrent I/O operations or for computationally intensive tasks that can be 
broken down into smaller, independent units of work.

The threading module is used to handle threads in Python. This module provides a way to create and manage threads in a Python program.
It allows us to create new threads, start them, and synchronize their execution. The threading module also provides several synchronization 
primitives, such as locks, events, and conditions, which can be used to coordinate the execution of multiple threads.
```

```python
 Explain the following functions?
run()
start()
join()
isAlive()
Ans:
    run(): This function is called when a thread is started using the start() method. 
    It contains the code that the thread will execute. This function should be overridden in a
    subclass of Thread to provide the behavior for the thread.

    start(): This function starts a thread by calling the run() method in a separate thread of execution. 
    The start() method should only be called once for each thread object. If the start() method is called 
    more than once on the same thread object, a RuntimeError will be raised.

    join(): This function blocks the calling thread until the thread being joined completes its execution.
    The optional timeout parameter specifies how long the calling thread should wait for the thread being joined to complete.
    If the timeout parameter is not specified, the join() function blocks until the thread being joined completes.

    isAlive(): This function returns a boolean value indicating whether the thread is currently running or not.
    If the thread is currently running, isAlive() returns True, otherwise it returns False. This function can be useful
    for determining whether a thread has completed its execution or not.
```

```python
# write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes
# Ans:
import threading

# Function to print the squares of numbers
def print_squares():
    for i in range(1, 11):
        print(i*i)

# Function to print the cubes of numbers
def print_cubes():
    for i in range(1, 11):
        print(i*i*i)

# Create two threads to print squares and cubes
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

# Start the threads
t1.start()
t2.start()

# Wait for both threads to complete
t1.join()
t2.join()
```

```python
State advantages and disadvantages of multithreading.
Ans:
Advantages:

    Improved performance: Multithreading can improve the performance of a program by utilizing multiple cores in a CPU 
    and reducing the waiting time for I/O operations. This can result in faster execution of the program.

    Increased responsiveness: Multithreading can make a program more responsive by allowing it to continue processing user 
    input while it is executing time-consuming tasks in the background.

    Resource sharing: Multithreading allows multiple threads to share resources such as memory, files, and network connections,
    which can result in more efficient use of system resources.

    Modularity: Multithreading allows the program to be organized into smaller, more manageable parts. Each thread can be
    responsible for a specific task, making the program easier to design, implement, and maintain.

Disadvantages:

    Complexity: Multithreading can make a program more complex to design, implement, and debug. The interactions between 
    threads can be difficult to manage, and it can be hard to ensure that the program behaves as expected.

    Synchronization: Multithreading requires synchronization between threads to ensure that shared resources are accessed 
    in a consistent and safe manner. This can add overhead to the program and increase the risk of deadlocks or race conditions.

    Resource contention: Multithreading can lead to resource contention if multiple threads try to access the same resource at
    the same time. This can result in slower execution and increased overhead.

    Debugging: Debugging multithreaded programs can be more difficult, as the interactions between threads can be hard to trace 
    and reproduce.
```

```python
 Explain deadlocks and race conditions
Ans:
    Deadlock:
A deadlock occurs when two or more threads are blocked, waiting for each other to release resources that they need to proceed. 
This can happen when multiple threads try to acquire the same resources in a different order, or when one thread acquires a 
resource and then waits for another resource that is held by a different thread.

For example, consider two threads, A and B, that need to access two resources, X and Y, in the following order: A needs to 
acquire X and then Y, while B needs to acquire Y and then X. If A acquires X and then waits for Y, and B acquires Y and then 
waits for X, a deadlock occurs because neither thread can proceed without releasing the resource that the other thread is waiting for.

Deadlocks can be difficult to detect and debug, and can lead to a program that appears to be frozen or unresponsive.

Race Condition:
A race condition occurs when two or more threads access a shared resource in an unpredictable order, which can lead to
unexpected or incorrect results. Race conditions occur when the order of access to the shared resource is not controlled
or synchronized properly.

For example, consider two threads, A and B, that both access a shared variable x. If both threads read the value of x, modify 
it, and then write it back, the final value of x depends on the order in which the threads execute. If thread A reads x first,
modifies it, and then writes it back, and thread B does the same thing, the final value of x will depend on the order in which 
the threads executed.

Race conditions can be difficult to detect and reproduce, and can lead to subtle bugs and unpredictable behavior in the program.
```
