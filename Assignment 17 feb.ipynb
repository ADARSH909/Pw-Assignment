```python
What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases
Ans:
MongoDB is a popular document-oriented NoSQL database system that is designed for storing, retrieving and managing large amounts of unstructured data.
In MongoDB, data is stored in a flexible document format called BSON, which is a binary representation of JSON.

Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional relational model of tables with rows and columns.
Instead, they use a variety of data models, including document-oriented, key-value, column-family, and graph-based models, to store and manage data. 
Non-relational databases are designed to handle large amounts of unstructured or semi-structured data and are highly scalable and flexible.

There are several scenarios where MongoDB may be preferred over SQL databases:

    Large volumes of unstructured or semi-structured data: MongoDB is designed to handle large volumes of unstructured or semi-structured data,
    making it an ideal choice for applications that generate a lot of data that does not fit neatly into tables and columns.

    High availability and scalability: MongoDB is designed to provide high availability and scalability, making it a good choice for applications
    that require a large amount of data to be stored and processed quickly.

    Agile development: MongoDB's flexible schema makes it well-suited for agile development and rapid prototyping. Changes to the data model can
    be made quickly and easily, without the need for complex database schema migrations.

    Real-time analytics: MongoDB's rich query language and support for real-time data aggregation make it a good choice for applications that 
    require real-time analytics, such as those in the financial, healthcare, and IoT industries.

    Cloud-native applications: MongoDB is designed to be cloud-native, making it a good choice for applications that are deployed in the cloud 
    and require high availability, scalability, and flexibility.
```

```python
State and Explain the features of MongoDB.
Ans:
    Document-oriented: MongoDB is a document-oriented database, which means that data is stored as flexible, JSON-like documents 
    instead of tables with rows and columns. This allows for greater flexibility and easier data modeling.

    Schema-less: MongoDB is a schema-less database, which means that there is no predefined schema for the data. This allows for
    greater flexibility in data modeling and makes it easier to make changes to the data model as needed.

    Indexing: MongoDB supports indexing of fields in documents, which makes it easier and faster to search and retrieve data.

    Aggregation: MongoDB supports a powerful aggregation framework that allows for real-time analysis of data and aggregation of 
    data across multiple collections.

    High availability and scalability: MongoDB is designed to provide high availability and scalability, with built-in replication 
    and sharding features that allow for automatic failover and load balancing.

    Security: MongoDB supports authentication and authorization, as well as encryption of data in transit and at rest, making it a 
    secure choice for storing sensitive data.

    Integration with other tools: MongoDB integrates with a wide range of other tools and technologies, including popular programming
    languages like Python and Java, as well as data visualization and analytics tools like Tableau and Power BI.

    Cloud-native: MongoDB is designed to be cloud-native, with features like automatic provisioning, scaling, and backup in popular 
    cloud platforms like AWS and Azure.
```

```python
Using the database and the collection created in question number 3, write a code to insert one record, 
and insert many records. Use the find() and find_one() methods to print the inserted record?
Ans:
import pymongo

# create a connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# create a new database and collection
mydb = client["mydatabase"]
mycol = mydb["customers"]

# insert one record into the collection
mydict = { "name": "John", "address": "Highway 37" }
x = mycol.insert_one(mydict)
print("Inserted one record:", x.inserted_id)

# insert multiple records into the collection
mylist = [
  { "name": "Amy", "address": "Apple st 652" },
  { "name": "Hannah", "address": "Mountain 21" },
  { "name": "Michael", "address": "Valley 345" },
  { "name": "Sandy", "address": "Ocean blvd 2" },
  { "name": "Betty", "address": "Green Grass 1" },
  { "name": "Richard", "address": "Sky st 331" },
  { "name": "Susan", "address": "One way 98" },
  { "name": "Vicky", "address": "Yellow Garden 2" },
  { "name": "Ben", "address": "Park Lane 38" },
  { "name": "William", "address": "Central st 954" },
  { "name": "Chuck", "address": "Main Road 989" },
  { "name": "Viola", "address": "Sideway 1633" }
]
y = mycol.insert_many(mylist)
print("Inserted multiple records:", y.inserted_ids)

# retrieve the inserted record using find_one()
result = mycol.find_one()
print("Inserted record:", result)

# retrieve all inserted records using find()
results = mycol.find()
print("All inserted records:")
for result in results:
    print(result)
```

```python
Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.
The find() method is used to query a MongoDB database to retrieve documents that match a specified set of criteria.
The find() method returns a cursor object that can be used to iterate over the retrieved documents.

Here's an example code that shows how to use the find() method to query a MongoDB database in Python:

import pymongo

# create a connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# select the database and collection
mydb = client["mydatabase"]
mycol = mydb["customers"]

# query the collection using the find() method
query = { "address": "Park Lane 38" }
results = mycol.find(query)

# print the results
for result in results:
    print(result)
```

```python
Explain the sort() method. Give an example to demonstrate sorting in MongoDB
Ans:
The sort() method is used to sort the results of a MongoDB query. The method can take a parameter that specifies the field to sort 
on, as well as the direction of the sort (ascending or descending).

Here's an example code that demonstrates how to use the sort() method in MongoDB:
import pymongo

# create a connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# select the database and collection
mydb = client["mydatabase"]
mycol = mydb["customers"]

# sort the collection by name in ascending order
results = mycol.find().sort("name")

# print the sorted results
for result in results:
    print(result)
```

```python
Explain why delete_one(), delete_many(), and drop() is used?
Ans:
In MongoDB, the delete_one() and delete_many() methods are used to remove one or multiple documents that match a specified set of criteria from
a collection. The drop() method is used to remove an entire collection from a database.

Here are some scenarios where each method might be used:

    delete_one(): This method is used to delete a single document that matches a specified set of criteria. For example, if you
    want to delete a customer document from a "customers" collection with a particular email address, you can use the delete_one()
    method with a query that matches the email address.

    delete_many(): This method is used to delete multiple documents that match a specified set of criteria. For example, if you want 
    to delete all customer documents from a "customers" collection with a status of "inactive", you can use the delete_many() method
    with a query that matches the status.

    drop(): This method is used to delete an entire collection from a database. For example, if you no longer need the "customers"
    collection, you can use the drop() method to remove it from the database.
```
